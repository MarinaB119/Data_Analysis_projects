```python
import pandas as pd
```

```python
df = pd.read_csv('../data/all_years_baby_names.csv')
```

```python
df.head()
```

```python
df_Emma = df[df['name']== 'Emma']
```

```python
df_Emma
```

```python
df_Emma_F = df_Emma[(df_Emma['gender'] == 'F') & (df_Emma['year']>1985)]
```

```python
df_Emma_F
```

```python
df_Emma_F.to_csv('../data/df_Emma_F.csv')
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize = (12,8))
plt.plot(df_Emma_F['year'], df_Emma_F['frequency'], label = 'Track Emma over time', color = "skyblue", marker= 'o', linestyle = '-')
plt.grid(True, linestyle='--', alpha=0.7)
plt.title("Name Frequency of Emma (1985-2020)")
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.legend()


```

```python
df_Joey = df[df['name']== 'Joey']
```

```python
df_Joey_M = df_Joey[(df_Joey['gender'] == 'M') & (df_Joey['year']>1985)]
```

```python
plt.figure(figsize = (12,8))
plt.plot(df_Joey_M['year'], df_Joey_M['frequency'], label = 'Track Joey over time', color = "royalblue", marker = 'o', linestyle = '-')
plt.grid(True, linestyle='--', alpha=0.7)
plt.title("Name Frequency of Joey (1985-2020)")
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.legend()

# Add a background color to the plot area for better contrast
plt.gca().set_facecolor('#f0f0f0')
```

```python
df_Ross = df[df['name']== 'Ross']
```

```python
df_Ross_M = df_Ross[(df_Ross['gender'] == 'M') & (df_Ross['year']>1985)]
```

```python
plt.figure(figsize = (12,8))
plt.plot(df_Ross_M['year'], df_Ross_M['frequency'], label = 'Track Ross over time', color = "navy", marker = 'd', linestyle= '--')

plt.grid(True, linestyle='--', alpha=0.7)
# Customize tick marks on the x-axis for better readability
plt.gca().set_facecolor('#f0f0f0')
plt.legend(shadow =True)
plt.title("Name Frequency of Ross (1985-2020)")
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.legend()

plt.xticks(range(1985, 2021, 10), rotation=45)
```
